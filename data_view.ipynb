```python
import h5py

# Specify the path to your .h5 file
h5_file_path = 'cataract_detection_vgg19.h5'  # Replace with the path to your .h5 file

# Open the .h5 file
with h5py.File(h5_file_path, 'r') as h5_file:
    # List all groups and datasets in the file
    def print_h5_items(name, obj):
        if isinstance(obj, h5py.Group):
            print(f'Group: {name}')
        elif isinstance(obj, h5py.Dataset):
            print(f'Dataset: {name}')

    h5_file.visititems(print_h5_items)

    # You can access specific datasets and their attributes using h5_file['dataset_name']
    # For example:
    dataset = h5_file['model_weights/block1_conv1/block1_conv1/bias:0']  # Replace 'dense_1' with the dataset name you want to access

    # Read and display dataset attributes
    for key, value in dataset.attrs.items():
        print(f'Attribute: {key} - Value: {value}')

    # Read and display dataset values (data)
    data = dataset[:]
    print(f'Dataset values (data):\n{data}')
```
